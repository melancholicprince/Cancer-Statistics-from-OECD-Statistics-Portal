```python
import numpy as np
import matplotlib.pyplot as plt

# here we put width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

incidence = [100, 83, 84, 42, 21, 32]
mortality = [20, 19, 22, 17, 12, 20]

br1 = np.arange(len(incidence))
br2 = [x + barWidth for x in br1]

# Makeing the plot
plt.bar(br1, incidence, color ='r', width = barWidth,
		edgecolor ='grey', label ='incidence')
plt.bar(br2, mortality, color ='g', width = barWidth,
		edgecolor ='grey', label ='mortality')


# Adding Xticks
plt.xlabel('Continents', fontweight ='bold', fontsize = 15)
plt.ylabel('Age-standartized per 100,000 persons', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(incidence))],
		['North America', 'Australia', 'Western Europe', 'Eastern Europe', 'Asia','Africa'])
plt.title("Breast cancer incidence and mortality rates",fontweight ='bold', fontsize = 28)

plt.grid(True)
plt.legend()
plt.show()
```
